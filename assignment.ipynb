## Question 1: matplotlib Fundamentals

### Part 1.1: Basic Figures and Subplots

**TODO: Create a figure with 2x2 subplots showing different plot types**

```python
# Create sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x/5)
y4 = np.log(x + 1)

# TODO: Create a 2x2 subplot grid
# TODO: Add titles and labels to each subplot
# TODO: Add grid to each subplot
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# TODO: Plot on each subplot
# Top-left: Sine wave
axes[0, 0].plot(x, y1, color = 'blue', linewidth = 2)
axes[0, 0].set_title('Sine Wave')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('sin(x)')
axes[0, 0].grid(True)

# Top-right: Cosine wave  
axes[0, 1].plot(x, y2, color = 'red', linewidth =2)
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('cos(x)')
axes[0, 1].grid(True)

# Bottom-left: Exponential decay
axes[1, 0].plot(x, y3, color = 'green', linewidth = 2)
axes[1, 0].set_title('Exponential Decay')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('exp(-x/5)')
axes[1, 0].grid(True)

# Bottom-right: Logarithmic function
axes[1, 1].plot(x, y4, color = 'magenta', linewidth =2)
axes[1, 1].set_title('Logarithmic Function')
axes[1, 1].set_xlabel('x')
axes[1, 1].set_ylabel('log(x+1)')
axes[1, 1].grid(True)


# TODO: Use tight_layout() to prevent overlapping
plt.tight_layout()
plt.show()
```

### Part 1.2: Plot Customization

**TODO: Create a customized line plot with multiple series**

```python
# Create sample data
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# TODO: Create a figure with custom styling
fig, ax = plt.subplots(figsize=(10, 6))

# TODO: Plot three lines with different styles
# Line 1: Solid line, blue color, label 'sin(x)'
# Line 2: Dashed line, red color, label 'cos(x)'  
# Line 3: Dotted line, green color, label 'sin(x)*cos(x)'
ax.plot(x, y1, linestyle = '-', color = 'blue', linewidth = 2, label = 'sin(x)')
ax.plot(x, y2, linestyle = '--', color = 'red', linewidth = 2, label = 'cos(x)')
ax.plot(x, y3, linestyle = ':', color = 'green', linewidth = 2, label = 'sin(x)*cos(x)')

# TODO: Add title, xlabel, ylabel
ax.set_title('Trigonometric Functions', fontsize = 16, fontweight = 'bold')
ax.set_xlabel('x', fontsize = 12)
ax.set_ylabel('y', fontsize = 12)

# TODO: Add legend
ax.legend(loc = 'upper right', fontsize = 10)

# TODO: Add grid with transparency
ax.grid(True, alpha = 0.3)

# TODO: Set appropriate axis limits
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)

plt.show()
```

### Part 1.3: Different Plot Types

```python
# Create sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 12, 78]
x_scatter = np.random.randn(100)
y_scatter = 2 * x_scatter + np.random.randn(100)
hist_data = np.random.normal(0, 1, 1000)

# TODO: Create a 2x2 subplot grid
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# TODO: Top-left: Bar chart
axes[0, 0].bar(categories, values, color = ['blue', 'red', 'orange', 'purple'], 
               edgecolor = 'black', linewidth = 1.5)
axes[0, 0].set_title('Bar Chart', fontsize = 14, fontweight = 'bold')
axes[0, 0].set_xlabel('Categories', fontsize = 12)
axes[0, 0].set_ylabel('Values', fontsize = 12)
axes[0, 0].grid(True, alpha = 0.3, axis = 'y')

# TODO: Top-right: Scatter plot
axes[0, 1].scatter(x_scatter, y_scatter, c = 'blue', alpha = 0.6,
                   edgecolors = 'black', linewidth = 0.5, s= 50)
axes[0, 1].set_title('Scatter Plot', fontsize = 14, fontweight = 'bold')
axes[0, 1].set_xlabel('x-values', fontsize = 12)
axes[0, 1].set_ylabel('y-values', fontsize = 12)
axes[0, 1].grid(True, alpha = 0.3)

# TODO: Bottom-left: Histogram
axes[1, 0].hist(hist_data, bins = 30, color = 'teal', alpha = 0.7, 
                edgecolor = 'black', linewidth = 1)
axes[1, 0].set_title('Histogram', fontsize = 14, fontweight = 'bold')
axes[1, 0].set_xlabel('Values', fontsize = 12)
axes[1, 0].set_ylabel('Frequency', fontsize = 12)
axes[1, 0].grid(True, alpha = 0.3, axis = 'y')

# TODO: Bottom-right: Pie chart
colors_pie = ['gold', 'lightblue', 'lightcoral', 'lightskyblue', 'lightgreen']
axes[1, 1].pie(values, labels = categories, colors = colors_pie, autopct = '%1.1f%%',
               startangle = 90, wedgeprops = {'edgecolor': 'black', 'linewidth': 1.5})
axes[1, 1].set_title('Pie Chart', fontsize = 14, fontweight='bold')

# TODO: Add appropriate titles and labels
# TODO: Customize colors and styles
# TODO: Save the plot as 'output/q1_matplotlib_plots.png'

plt.tight_layout()
plt.savefig('output/q1_matplotlib_plots.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Question 2: seaborn Statistical Visualization

### Part 2.1: Load and Explore Data

```python
# Load the sales data
sales_df = pd.read_csv('data/sales_data.csv')
customer_df = pd.read_csv('data/customer_data.csv')
product_df = pd.read_csv('data/product_data.csv')

# Merge data for analysis
merged_df = sales_df.merge(customer_df, on='customer_id').merge(product_df, on='product_id')

print("Data shape:", merged_df.shape)
print("\nColumns:", merged_df.columns.tolist())
print("\nFirst few rows:")
print(merged_df.head())
```

### Part 2.2: Statistical Plots

**TODO: Create statistical visualizations with seaborn**

```python
# TODO: Create a figure with 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# TODO: Top-left: Box plot of total_amount by store_location
sns.boxplot(data = merged_df, x = 'store_location', y = 'total_amount', 
            palette = 'Set2', ax = axes[0, 0])
axes[0, 0].set_title('Total Amount by Store Location', fontsize = 14, fontweight ='bold')
axes[0, 0].set_xlabel('Store Location', fontsize = 12)
axes[0, 0].set_ylabel('Total Amount ($)', fontsize = 12)
axes[0, 0].grid(True, alpha = 0.3, axis = 'y')

# TODO: Top-right: Violin plot of total_amount by category
sns.violinplot(data = merged_df, x = 'category', y = 'total_amount', 
               palette = 'muted', ax = axes[0, 1])
axes[0, 1].set_title('Total Amount Distribution by Category', fontsize = 14, fontweight = 'bold')
axes[0, 1].set_xlabel('Category', fontsize = 12)
axes[0, 1].set_ylabel('Total Amount ($)', fontsize = 12)
axes[0, 1].tick_params(axis = 'x', rotation = 45)
axes[0, 1].grid(True, alpha = 0.3, axis = 'y')

# TODO: Bottom-left: Scatter plot of quantity vs total_amount with hue by gender
sns.scatterplot(data = merged_df, x = 'quantity', y = 'total_amount', 
                hue='gender', palette = 'viridis', alpha = 0.6, s = 80, ax = axes[1, 0])
axes[1, 0].set_title('Quantity vs Total Amount by Gender', fontsize = 14, fontweight = 'bold')
axes[1, 0].set_xlabel('Quantity', fontsize = 12)
axes[1, 0].set_ylabel('Total Amount ($)', fontsize = 12)
axes[1, 0].legend(title = 'Gender', fontsize = 10)
axes[1, 0].grid(True, alpha = 0.3)

# TODO: Bottom-right: Histogram of total_amount with kde overlay
sns.histplot(data = merged_df, x = 'total_amount', bins = 30, kde = True, 
             color = 'skyblue', edgecolor = 'black', linewidth = 1, ax = axes[1, 1])
axes[1, 1].set_title('Distribution of Total Amount', fontsize = 14, fontweight = 'bold')
axes[1, 1].set_xlabel('Total Amount ($)', fontsize = 12)
axes[1, 1].set_ylabel('Frequency', fontsize = 12)
axes[1, 1].grid(True, alpha = 0.3, axis = 'y')

# TODO: Add appropriate titles and labels
# TODO: Customize colors and styling
# TODO: Save the plot as 'output/q2_seaborn_plots.png'

plt.tight_layout()
plt.savefig('output/q2_seaborn_plots.png', dpi=300, bbox_inches='tight')
plt.show()
```
